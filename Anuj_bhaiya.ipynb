# Operators In Python

```python
#Arithmetic Operator
#Average Of Three Numbers
first_num  = 45
second_num = 23
third_num  = 13
average=(first_num+second_num +third_num/3)
print(average)
```

```python
#Assignment Operator
a = 6
#a = a + 56
a += 56
print (a)
```

```python
#Relational Operators
a = 56
b = 78

print("a > b",a>b)
print("a < b",a<b)
print("a == b",a==b)
print("a != b",a!=b)
print("a >= b",a>=b)
print("a <= b",a<=b)


```

```python
#Logical Oprator and or not
x = True
y = False

print("x and y",x and y)
print("x or y",x or y)
print("x not ", not x )
```

```python
#if statement
age = 23
if age > 18:
    print('He can Vote')
```

```python
#if else
age = 15
if age > 18:
    print('You can Vote')
else:
    print('You Can not Vote')
```

```python
#elif
num = 6.5
if num>0:
     print("number is positive")
elif num == 0:
     print("number is zero")
else:
    print("number is negative")  
```

# Break and Continue

```python
#break
for i in list(range(20)):
    if i10:
      break
    print(i)
```

```python
#Continue
for i in list(range(20)):
    if i==10:
      continue
    print(i)
```
